### 1. La importancia de la visualización en el análisis exploratorio de datos <a name="id1"></a>


La visualización tiene un papel fundamental en el análisis exploratorio de datos (EDA) y es uno de los primeros pasos en la exploración de un dataset.

Cuando representamos información de manera visual podemos identificar tendencias o relaciones entre variables de manera más rápida, así como detectar anomalías en los datos. Una buena visualización también ayuda a escoger de manera más efectiva los análisis que se realizan posteriormente, ya que permite comprobar la distribución de los datos (e.g. condición de normalidad) o indicar la presencia de agrupaciones o patrones no lineales.

La visualización no es solo útil al inicio del análisis, sino también al final, ya que es un elemento básico para comunicar los resultados de manera entendedora, tanto en audiencias técnicas como no técnicas. Las personas tienden a procesar la información visual de manera más intuitiva y rápida que los datos en forma de tablas o números, por eso una visualización adecuada hace que la información sea más accesible.

### 2. Gráficos más comunes y cuando usarlos <a name="id2"></a>

Existen diversos tipos de gráficos, cada uno diseñado para representar diferentes tipos de datos y patrones.

Es crucial seleccionar el tipo de gráfico adecuado para los datos específicos que estás tratando de visualizar. La elección incorrecta de gráfico puede llevar a malinterpretaciones o a una representación ineficaz de los datos.

A continuación, vamos a ver ejemplos de los gráficos más comunes.

#### 2.1 Histograma

Un histograma es una representación gráfica de la distribución de un conjunto de datos. Este tipo de gráfico divide el rango de valores en intervalos (o "bins") y muestra la frecuencia con la que ocurren los datos dentro de cada intervalo.

**Cuándo usarlo:** Para representar la distribución de una variable numérica y visualizar la frecuencia de rangos o clases.

```python
# Cargamos las librerías necesarias
import matplotlib.pyplot as plt
import numpy as np

# Generamos unos datos de ejemplo
# Son datos distribuidos normalmente
datos = np.random.randn(1000)

# Creamos el histograma
plt.hist(datos, bins=30, color='blue', alpha=0.7)
plt.title('Ejemplo de Histograma')
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.show()
```

#### 2.2 Gráfico de Barras

Un gráfico de barras es una representación visual de datos que utiliza barras rectangulares para mostrar la comparación entre diferentes categorías o grupos. La longitud de cada barra es proporcional a la magnitud de la variable que representa.

**Cuándo usarlo**: Ideal para representar datos categóricos o comparar frecuencias de diferentes categorías.

```python
# Cargamos las librerías necesarias
import matplotlib.pyplot as plt

# Creamos unos datos de ejemplo
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [15, 24, 10, 30, 18]

# Creamos gráfico de barras
plt.bar(categorias, valores, color='skyblue')
plt.title('Ejemplo de Gráfico de Barras')
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.show()
```

#### 2.3 Gráfico de Línea

Un gráfico de línea es una representación visual de datos en la que se utilizan líneas para conectar puntos de datos en un espacio bidimensional. Este tipo de gráfico es efectivo para mostrar la tendencia o cambio en los datos a lo largo de un eje continuo, como el tiempo.

**Cuándo usarlo**: Adecuado para mostrar cambios a lo largo de un eje continuo (como el tiempo) o secuencia. Útil para visualizar tendencias.

```python
# Cargamos las librerías necesarias
import matplotlib.pyplot as plt
import numpy as np

# Generamos datos de ejemplo
x = np.linspace(0, 10, 100)  # Valores en el eje x
y = np.sin(x)  # Valores en el eje y (en este caso, el seno de x)

# Creamos gráfico de línea
plt.plot(x, y, label='Función Seno', color='green', linestyle='--', linewidth=2)

# Añadimos etiquetas y leyenda
plt.title('Ejemplo de Gráfico de Línea')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.legend()

# Mostramos el gráfico
plt.show()
```

#### 2.4 Gráfico de Dispersión


Un gráfico de dispersión es una representación visual de datos en la que se utilizan puntos para representar las coordenadas de dos variables. Cada punto en el gráfico de dispersión representa una observación y se coloca en el espacio bidimensional según sus valores en las dos variables.

**Cuándo usarlo**: Para explorar la relación entre dos variables numéricas. Útil para identificar patrones, correlaciones y outliers.

```python
# Cargamos las librerías necesarias
import matplotlib.pyplot as plt
import numpy as np

# Generamos datos de ejemplo
x = np.random.rand(50)  # Variable X aleatoria
y = 2 * x + 1 + 0.1 * np.random.randn(50)  # Variable Y relacionada con X con ruido

# Creamos el gráfico de dispersión
plt.scatter(x, y, color='blue', alpha=0.7)

# Añadimos etiquetas y título
plt.title('Ejemplo de Gráfico de Dispersión')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')

# Mostramos el gráfico
plt.show()
```

#### 2.5 Diagrama de Caja (Boxplot)

Un diagrama de caja (boxplot) es una representación gráfica que proporciona un resumen de varias características importantes de un conjunto de datos, como la mediana, los cuartiles y los valores atípicos.

Las partes principales del diagrama de caja son:

- La caja: Representa el rango intercuartílico (IQR), que cubre el 50% central de los datos.
- La línea dentro de la caja: Representa la mediana.
- Los bigotes: Representan la extensión de los datos, excluyendo los valores atípicos.
- Puntos individuales (outliers): Representan valores atípicos que caen fuera de los bigotes.

**Cuándo usarlo**: Para visualizar la distribución, mediana, cuartiles y detectar valores atípicos en una variable numérica.

```python
# Cargamos las librerías necesarias
import matplotlib.pyplot as plt
import numpy as np

# Generamos datos de ejemplo
np.random.seed(42)
datos = np.random.randn(100)

# Creamos el diagrama de caja
# patch_artist llena las cajas con color
plt.boxplot(datos, patch_artist=True)

# Añadimos título y etiquetas
plt.title('Ejemplo de Diagrama de Caja')
plt.xlabel('Valores')

# Mostramos el diagrama de caja
plt.show()
```

#### 2.6 Diagrama de Violín


Un diagrama de violín es una representación gráfica que combina elementos de un boxplot (diagrama de caja) y un kernel density plot. Muestra la distribución de datos a lo largo de un eje, permitiendo visualizar la forma de la distribución, la concentración de datos y la presencia de múltiples modas.

El diagrama de violín resultante muestra una "caja" en el centro, que es similar al rango intercuartílico (IQR) en un boxplot, y "violines" a cada lado que representan la densidad de los datos en esa ubicación a lo largo del eje y.

**Cuándo usarlo**: Este tipo de visualización es útil cuando se desea comparar la distribución de datos entre diferentes categorías. El ancho del violín indica la densidad de datos en diferentes regiones del eje y, y se puede utilizar para detectar modas y asimetrías en la distribución.

```python
# Cargamos las librerías necesarias
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generamos datos de ejemplo
categoria1 = np.random.normal(0, 1, 100)
categoria2 = np.random.normal(3, 1, 100)

# Creamos diagrama de violín
sns.violinplot(
    x=['Categoría 1']*100 + ['Categoría 2']*100,
    y=np.concatenate([categoria1, categoria2])
    )

# Añadimos título y etiquetas
plt.title('Ejemplo de Diagrama de Violín')
plt.xlabel('Categoría')
plt.ylabel('Valores')

# Mostramos el diagrama de violín
plt.show()
```

#### 2.7 Gráfico circular (Pie chart)


Un gráfico circular (o pie chart en inglés) es una representación gráfica que muestra proporciones en un conjunto de datos, dividiendo un círculo en segmentos que representan diferentes partes del todo. Cada segmento está etiquetado y su tamaño es proporcional al porcentaje que representa en relación con el total.


**Cuándo usarlo**: Adecuado para representar proporciones de un todo. Sin embargo, su uso se debe limitar a un pequeño número de categorías para evitar confusiones.

```python
# Cargamos las librerías necesarias
import matplotlib.pyplot as plt

# Datos de ejemplo
porcentajes = [30, 20, 15, 35]
categorias = ['Categoría A', 'Categoría B', 'Categoría C', 'Categoría D']

# Creamos gráfico circular
plt.pie(porcentajes, labels=categorias, autopct='%1.1f%%', startangle=90,
        colors=['lightblue', 'lightgreen', 'lightcoral', 'lightsalmon'])

# Añadimos título
plt.title('Ejemplo de Gráfico Circular')

# Mostramos el gráfico circular
plt.show()
```

#### 2.8 Mapa de Calor (Heatmap)

Un mapa de calor (heatmap) es una representación visual de datos en una matriz, donde los valores individuales se representan con colores. Cada celda en la matriz se colorea según el valor que contiene, permitiendo visualizar patrones, tendencias y distribuciones en datos bidimensionales.

**Cuándo usarlo**: Para visualizar la intensidad de una relación entre dos variables en una matriz. Comúnmente utilizado en análisis de correlación o para representar matrices.

```python
# Cargamos las librerías necesarias
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Generamos datos de ejemplo
datos = np.random.rand(10, 10)

# Creamos mapa de calor
sns.heatmap(datos, annot=True, cmap='viridis')

# Añadimos título
plt.title('Ejemplo de Mapa de Calor')

# Mostramos el mapa de calor
plt.show()
```

#### 2.9 Gráfico de Área

Un gráfico de área es una representación gráfica que muestra la evolución de una o más variables a lo largo del tiempo o a lo largo de una secuencia. Este tipo de gráfico utiliza áreas sombreadas bajo las curvas para representar la magnitud acumulativa de los valores en un eje, lo que facilita la comparación de tendencias entre las variables.


**Cuándo usarlo**: Similar al gráfico de líneas, pero útil cuando quieres resaltar la acumulación de valores a lo largo del tiempo. También se usa para mostrar la evolución de variables como ingresos, población, o cualquier medida que pueda acumularse o desglosarse en varias categorías.

```python
# Cargamos las librerías necesarias
import matplotlib.pyplot as plt
import numpy as np

# Generamos datos de ejemplo
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Creamos gráfico de área
plt.fill_between(x, y1, alpha=0.5, label='Área bajo y1', color='skyblue')
plt.fill_between(x, y2, alpha=0.5, label='Área bajo y2', color='lightcoral')

# Añadimos etiquetas y título
plt.title('Ejemplo de Gráfico de Área')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

# Añadimos leyenda
plt.legend()

# Mostramos el gráfico de área
plt.show()
```

#### 2.10 Gráfico de Burbujas


Un gráfico de burbujas es una representación visual de datos que utiliza burbujas para mostrar la relación entre tres variables. Cada burbuja en el gráfico representa una observación y tiene un tamaño proporcional al valor de una variable, mientras que la posición en los ejes x e y representa los valores de otras dos variables.

**Cuándo usarlo**: Para representar tres dimensiones de datos: dos en el eje y una mediante el tamaño de las burbujas.

```python
# Cargamos las librerías necesarias
import matplotlib.pyplot as plt
import numpy as np

# Generamos datos de ejemplo
np.random.seed(42)
x = np.random.rand(20)
y = np.random.rand(20)
tamanos = 100 * np.random.rand(20)  # Tamaños de las burbujas
colores = np.random.rand(20)  # Colores de las burbujas

# Creamos gráfico de burbujas
plt.scatter(x, y, s=tamanos, c=colores, alpha=0.7, cmap='viridis')

# Añadimos etiquetas y título
plt.title('Ejemplo de Gráfico de Burbujas')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

# Mostramos el gráfico de burbujas
plt.show()
```

#### 2.11 Gráfico de Radar

Un gráfico de radar, también conocido como gráfico de araña o gráfico de estrella, es una representación visual que muestra datos en varias categorías representadas en un formato circular. Cada categoría se representa por un eje radial y los valores se marcan a lo largo de cada eje, creando una figura poligonal que conecta los puntos.

**Cuándo usarlo**: Este tipo de visualización es útil para comparar valores en múltiples categorías y resaltar patrones o desviaciones en los datos. También se usa para resaltar fortalezas y debilidades en diferentes aspectos.

```python
# Cargamos las librerías necesarias
import matplotlib.pyplot as plt
import numpy as np

# Generamos datos de ejemplo
categorias = ['Categoría 1', 'Categoría 2', 'Categoría 3', 'Categoría 4', 'Categoría 5']
valores = [4, 3, 5, 2, 4]

# Añadimos la primera categoría al final para cerrar el gráfico
categorias += [categorias[0]]
valores += [valores[0]]

# Calculamos los ángulos para los ejes radiales
angulos = np.linspace(0, 2 * np.pi, len(categorias), endpoint=False)

# Creamos gráfico de radar
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angulos, valores, color='skyblue', alpha=0.7)

# Añadimos etiquetas
ax.set_xticks(angulos)
ax.set_xticklabels(categorias)

# Añadimos título
plt.title('Ejemplo de Gráfico de Radar')

# Mostramos el gráfico de radar
plt.show()
```

#### 2.12 Gráfico de Pareto


Un gráfico de Pareto es un tipo de gráfico de barras que combina datos de barras ordenadas de mayor a menor con una línea que muestra la acumulación total. Este gráfico es utilizado para identificar y priorizar las causas o problemas más significativos dentro de un conjunto de datos.

El principio de Pareto, también conocido como la regla 80/20, sugiere que aproximadamente el 80% de los efectos provienen del 20% de las causas.

**Cuándo usarlo**: Para identificar visualmente las causas más significativas y la contribución acumulativa de estas causas al total. Este tipo de visualización es útil en diversas áreas, como gestión de calidad, análisis de negocios y resolución de problemas, para tomar decisiones basadas en la priorización de las causas más relevantes.

```python
# Cargamos las librerías necesarias
import matplotlib.pyplot as plt

# Datos de ejemplo
causas = ['Causa A', 'Causa B', 'Causa C', 'Causa D', 'Causa E']
frecuencias = [25, 20, 15, 10, 30]

# Ordenamos las causas y frecuencias de mayor a menor
datos_ordenados = sorted(zip(causas, frecuencias), key=lambda x: x[1], reverse=True)
causas_ordenadas, frecuencias_ordenadas = zip(*datos_ordenados)

# Calculamos la frecuencia acumulada
frecuencia_acumulada = [sum(frecuencias_ordenadas[:i+1]) for i in range(len(frecuencias_ordenadas))]

# Creamos gráfico de Pareto
fig, ax1 = plt.subplots()

# Barras
ax1.bar(causas_ordenadas, frecuencias_ordenadas, color='lightblue')
ax1.set_xlabel('Causas')
ax1.set_ylabel('Frecuencias', color='blue')
ax1.tick_params('y')

# Línea de frecuencia acumulada
ax2 = ax1.twinx()
ax2.plot(causas_ordenadas, frecuencia_acumulada, color='orange', marker='o')
ax2.set_ylabel('Frecuencia Acumulada', color='orange')

# Añadimos título
plt.title('Gráfico de Pareto')

# Mostramos el gráfico de Pareto
plt.show()
```

#### 2.13 Diagrama de rosquilla (Doughnut Chart)

El diagrama de rosquilla (o donut chart) es una variante del gráfico circular (pie chart) que presenta un agujero en el centro.

**Cuándo usarlo**:  A menudo se usa para resaltar una categoría específica o para mejorar la estética del gráfico, pero se debe usar con precaución.

```python
# Cargamos las librerías necesarias
import matplotlib.pyplot as plt

# Datos de ejemplo
categorias = ['Categoría A', 'Categoría B', 'Categoría C', 'Categoría D']
valores = [25, 20, 15, 40]

# Creamos el Doughnut Chart
fig, ax = plt.subplots()
ax.pie(valores, labels=categorias, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.4))

# Añadimos círculo en el centro para crear el agujero
centro_circulo = plt.Circle((0, 0), 0.25, fc='white')
ax.add_patch(centro_circulo)

# Ajustamos el aspecto del gráfico
ax.axis('equal')

# Añadimos título
plt.title('Ejemplo de Doughnut Chart')

# Mostramos el gráfico
plt.show()
```

### 3. Distribuciones de datos <a name="id3"></a>

La distribución de datos se refiere a la forma en que se distribuyen los valores en un conjunto de datos. Describe la frecuencia y la probabilidad de ocurrencia de diferentes valores en una variable. Comprender la distribución de datos es fundamental en estadísticas y análisis de datos, ya que puede proporcionar información valiosa sobre la tendencia central, la variabilidad y la forma de los datos.

En la práctica, se usan herramientas como histogramas, gráficos de probabilidad, y pruebas estadísticas para comprender y seleccionar la distribución que mejor se ajusta a los datos.

Aquí hay algunos tipos comunes de distribuciones de datos:

1. **Distribución Normal (Gaussiana):**
   - También conocida como campana de Gauss.
   - Es simétrica alrededor de la media.
   - Muchos fenómenos en la naturaleza siguen una distribución normal.

2. **Distribución Uniforme:**
   - Todos los valores tienen la misma probabilidad de ocurrir.
   - Se representa como una línea horizontal en un histograma.

3. **Distribución Exponencial:**
   - Modela el tiempo entre eventos en un proceso.
   - Tiene una cola larga hacia la derecha.
   - *Ejemplo*: el tiempo que tarda una partícula radiactiva en desintegrarse.



```python
# Cargamos las librerías necesarias
import numpy as np
import matplotlib.pyplot as plt

# Configuración de la figura
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Generamos datos con distribución Normal
mu, sigma = 0, 1  # Media y desviación estándar
datos_normal = np.random.normal(mu, sigma, 1000)
axs[0].hist(datos_normal, bins=30, color='skyblue', edgecolor='black', density=True)
axs[0].set_title('Distribución Normal')

# Generamos datos con distribución Uniforme
datos_uniforme = np.random.uniform(-1, 1, 1000)
axs[1].hist(datos_uniforme, bins=30, color='lightgreen', edgecolor='black', density=True)
axs[1].set_title('Distribución Uniforme')

# Generamos datos con distribución Exponencial
datos_exponencial = np.random.exponential(scale=1, size=1000)
axs[2].hist(datos_exponencial, bins=30, color='lightcoral', edgecolor='black', density=True)
axs[2].set_title('Distribución Exponencial')

# Ajustes comunes para todos los subgráficos
for ax in axs:
    ax.grid(True)
    ax.set_xlabel('Valor')
    ax.set_ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

4. **Distribución de Poisson:**
   - Modela la cantidad de eventos que ocurren en un intervalo de tiempo fijo.
   - Utilizado para eventos raros pero que ocurren de manera constante.

5. **Distribución Binomial:**
   - Modela el número de éxitos en un número fijo de ensayos independientes.
   - Cada ensayo tiene dos resultados posibles (éxito o fracaso).

6. **Distribución Gamma:**
   - Se utiliza en modelado de tiempos de espera y duraciones.
   - En teoría de colas y análisis de sistemas, la distribución gamma es útil para modelar el tiempo entre llegadas de eventos.


```python
# Cargamos las librerías necesarias
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson, binom, gamma

# Configuración de la figura
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Generamos datos con distribución de Poisson
mu_poisson = 3  # Parámetro lambda
datos_poisson = poisson.rvs(mu=mu_poisson, size=1000)
axs[0].hist(datos_poisson, bins=np.arange(0, 10), color='skyblue', edgecolor='black', density=True)
axs[0].set_title('Distribución de Poisson')

# Generamos datos con distribución Binomial
n_binomial, p_binomial = 10, 0.3  # Parámetros n y p
datos_binomial = binom.rvs(n=n_binomial, p=p_binomial, size=1000)
axs[1].hist(datos_binomial, bins=np.arange(0, 11), color='lightgreen', edgecolor='black', density=True)
axs[1].set_title('Distribución Binomial')

# Generamos datos con distribución de Gamma
a_gamma, scale_gamma = 2, 1  # Parámetros shape (a) y scale (theta)
datos_gamma = gamma.rvs(a=a_gamma, scale=scale_gamma, size=1000)
axs[2].hist(datos_gamma, bins=np.linspace(0, 10, 20), color='lightcoral', edgecolor='black', density=True)
axs[2].set_title('Distribución Gamma')

# Ajustes comunes para todos los subgráficos
for ax in axs:
    ax.grid(True)
    ax.set_xlabel('Valor')
    ax.set_ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

7. **Distribución de Pareto:**
   - Sigue la regla 80/20, donde el 80% de los efectos provienen del 20% de las causas.
   - Utilizado en economía y teoría de sistemas complejos.

8. **Distribución Log-Normal:**
   - Los logaritmos de las observaciones siguen una distribución normal.
   - Utilizado para modelar precios de activos financieros.

9. **Distribución de Weibull:**
   - Utilizado para modelar tiempos de vida y fiabilidad en ingeniería.

```python
# Cargamos las librerías necesarias
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pareto, lognorm, weibull_min

# Configuración de la figura
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Generamos distribución de Pareto
alpha_pareto = 3.0  # Parámetro de forma
datos_pareto = pareto.rvs(alpha_pareto, size=1000)
axs[0].hist(datos_pareto, bins=np.linspace(1, 10, 20), color='skyblue', edgecolor='black', density=True)
axs[0].set_title('Distribución de Pareto')

# Generamos distribución Log-Normal
mu_lognormal, sigma_lognormal = 0, 0.8  # Parámetros mu y sigma
datos_lognormal = lognorm.rvs(sigma_lognormal, scale=np.exp(mu_lognormal), size=1000)
axs[1].hist(datos_lognormal, bins=np.linspace(0, 5, 20), color='lightgreen', edgecolor='black', density=True)
axs[1].set_title('Distribución Log-Normal')

# Generamos distribución Weibull
c_weibull = 1.5  # Parámetro de forma
datos_weibull = weibull_min.rvs(c_weibull, size=1000)
axs[2].hist(datos_weibull, bins=np.linspace(0, 5, 20), color='lightcoral', edgecolor='black', density=True)
axs[2].set_title('Distribución Weibull')

# Ajustes comunes para todos los subgráficos
for ax in axs:
    ax.grid(True)
    ax.set_xlabel('Valor')
    ax.set_ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```
