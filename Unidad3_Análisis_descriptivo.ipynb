### 1. Estadísticos descriptivos <a name="id1"></a>

Los **estadísticos descriptivos** son medidas numéricas o gráficos que resumen y describen las características principales de un conjunto de datos. Estos proporcionan un resumen simple pero informativo de la distribución, dispersión y tendencias centrales de los datos.

Los estadísticos más comunes son los siguientes:

1. **Medidas de Tendencia Central:**
   - **Media (promedio):** Suma de todos los valores dividida por el número de observaciones.
   - **Mediana:** El valor que se encuentra en el centro de un conjunto de datos cuando está ordenado.
   - **Moda:** El valor que aparece con mayor frecuencia en un conjunto de datos.

2. **Medidas de Dispersión:**
   - **Rango:** La diferencia entre el valor máximo y mínimo en un conjunto de datos.
   - **Varianza:** La medida de la dispersión de los datos con respecto a la media.
   - **Desviación estándar:** La raíz cuadrada de la varianza, proporciona una medida de la dispersión en la misma unidad que los datos.

3. **Medidas de Posición:**
   - **Cuartiles:** Dividen el conjunto de datos en cuatro partes iguales.
   - **Percentiles:** Dividen el conjunto de datos en cien partes iguales.

4. **Medidas de Forma:**
   - **Asimetría:** Mide la simetría o la falta de simetría en una distribución de datos.
   - **Curtosis:** Mide la forma de las colas de la distribución de datos.

5. **Estadísticos Resumen:**
   - **Recuento de observaciones:** Número total de elementos en el conjunto de datos.
   - **Suma total:** La suma de todos los valores en el conjunto de datos.
   - **Porcentaje de valores faltantes/nulos (NaNs)**: Proporción de datos que faltan o son erróneos en el conjunto.

6. **Gráficos Descriptivos**:
  - **Histogramas**: Representación gráfica de la distribución de frecuencias.
  - **Diagrama de caja (boxplot)**: Muestra la distribución de los datos a través de cuartiles.
  - **Diagrama de dispersión**: Muestra la relación entre dos variables.

Como recordatorio, la **caja** representa el rango intercuartílico (IQR), que es la distancia entre el primer cuartil (Q1) y el tercer cuartil (Q3). La línea dentro de la caja indica la mediana.  Los **bigotes** se extienden desde los extremos de la caja hasta los valores extremos dentro de un rango específico (normalmente es 1.5 veces el IQR). Los valores fuera de este rango se consideran anómalos y se muestran como puntos individuales.

Por lo tanto, es importante identificar si hay valores anómalos en el dataset, ya que puede impactar en los estadísticos y dar lugar a interpretaciones erróneas!

Así pues, el histograma nos permite sacar información/conclusiones de manera muy visual!

Si queremos sacar esta información con estadísticos, podemos calcular la asimetría y la curtosis.

**Asimetría**

La asimetría mide la falta de simetría en una distribución.
- Una distribución simétrica tiene una asimetría igual a cero.
- Si la cola derecha es más larga que la izquierda (tiene más valores diferentes), la asimetría es positiva.
- Si la cola izquierda es más larga que la derecha, la asimetría es negativa.

**Curtosis**

La curtosis mide la "altura" y forma de las colas de una distribución en relación con la media.
- La curtosis positiva indica colas más pesadas y un pico más pronunciado.
- La curtosis negativa indica colas más ligeras y una distribución más aplanada.

```python
# Importamos la librería scripy con el módulo stats
from scipy.stats import skew, kurtosis

# Calculamos la asimetría y la kurtosis de la variable IMDb Rating
asimetria = skew(got_df["IMDb Rating"])
curtosis = kurtosis(got_df["IMDb Rating"])

# Mostramos los estadísticos
print("Asimetría:", round(asimetria, 2))
print("Curtosis:", round(curtosis, 2))
```

Si interpretamos los resultados vemos que la distribución no es simétrica y que tiene la cola izquierda más larga que la derecha (valor de simetría negativo). Por otro lado, un valor alto de kurtosis indica que las colas son largas y que hay un pico pronunciado, a diferencia de una distribución normal.

```python
# ¿Cómo se distribuyen las visualizaciones en US?

# Vamos a usar un histograma para ver la distribución de esta variable
sns.histplot(got_df['U.S. Viewers (Millions)'])
```

```python
# Importamos la librería scripy con el módulo stats
from scipy.stats import skew, kurtosis

# Calculamos la asimetría y la kurtosis de la variable U.S. Viewers (Millions)
asimetria2 = skew(got_df["U.S. Viewers (Millions)"])
curtosis2 = kurtosis(got_df["U.S. Viewers (Millions)"])

# Mostramos los estadísticos
print("Asimetría:", round(asimetria2, 2))
print("Curtosis:", round(curtosis2, 2))
```

A diferencia de la variable anterior, la distribución de las visualizaciones está sesgada a derecha (asimetría positiva). Por otro lado, el valor negativo de curtosis nos indica que la distribución es más aplanada que una distribución normal.

### 2. Gestión del tipo de variables <a name="id2"></a>

La gestión del tipo de variables en el análisis exploratorio de datos es esencial para realizar análisis estadísticos y visuales significativos.

Las variables pueden clasificarse en diferentes tipos y, según el tipo, se deben tratar y interpretar de manera diferente.

1. **Variables Numéricas:**
   - **Continuas:** Pueden tomar cualquier valor en un rango (por ejemplo, altura, gastos).
   - **Discretas:** Solo pueden tomar valores específicos (por ejemplo, puntuación de una evaluación).

   *Como tratarlas:*
   - Uso de estadísticas descriptivas (media, mediana, desviación estándar).
   - Visualización mediante histogramas, gráficos de dispersión y boxplots.

2. **Variables Categóricas:**
   - **Nominales:** No tienen un orden inherente (por ejemplo, color, género).
   - **Ordinales:** Tienen un orden pero no necesariamente una distancia constante entre categorías (por ejemplo, nivel de educación, clasificación).

   *Como tratarlas:*
   - Uso de gráficos de barras, diagramas circulares (pie chart) y tablas de frecuencia.
   - Análisis de tendencias y patrones.

3. **Variables Temporales:**
   - **Fecha y Hora:** Representan momentos específicos en el tiempo.

   *Como tratarlas:*
   - Uso de gráficos de series temporales.
   - Análisis de tendencias y estacionalidad.

4. **Variables Textuales:**
   - **Cadenas de Texto:** Descripciones, comentarios.

   *Como tratarlas:*
   - Análisis de frecuencia de palabras.
   - Procesamiento de lenguaje natural (NLP) para análisis más avanzados.

5. **Variables Binarias:**
   - **Sí/No o 0/1:** Representan información de booleanos.

   *Como tratarlas:*
   - Análisis de proporciones y porcentajes.
   - Uso de gráficos de barras o diagramas circulares (pie chart).

Por otro lado, también hemos visto que tenemos la variable `Ordered`, que nos da información sobre la fecha de emisión. Sin embargo, esta columna está en formato string, no en formato fecha.

```python
# Nos fijamos que Ordered aparece como object (string)
# pero debería ser una columna de fecha
# Vamos a ver sus valores
print("Los valores únicos de Ordered son:\n", got_df['Ordered'].unique())
```

Vamos a transformarla para tenerla en el formato correcto.

Para hacerlo vamos a usar el módulo `datetime`.

```python
# Importamos el módulo datetime
from datetime import datetime

# Cadena de fecha en formato original
cadena_fecha = "March 2, 2010"

# Utilizamos datetime.strptime para analizar la cadena de fecha
# %B representa el nombre del mes en formato largo (March)
# %d representa el número del dia (2)
# %Y representa el año en formato completo (2010)

# Podéis encontrar las equivalencias en https://www.w3schools.com/python/python_datetime.asp
fecha_objeto = datetime.strptime(cadena_fecha, "%B %d, %Y")
print("El elemento en formato fecha (date) es:\n", fecha_objeto)
```

```python
# Otro ejemplo

# Cadena de fecha y hora original
fecha_hora_str = "2023/12/03 15*30*00"

# %Y representa el año en formato completo (2013)
# %m representa el número del mes (12)
# %d representa el número del dia (03)
# %H representa la hora en formato 00-23 (15)
# %M representa los minutos en formato 00-59 (30)
# %S representa los segundos en formato 00-59 (00)

# Formato de la cadena
formato_str = "%Y/%m/%d %H*%M*%S"

# Usamos de strptime para convertir la cadena a un objeto de fecha y hora
fecha_hora_obj = datetime.strptime(fecha_hora_str, formato_str)

# Mostramos el resultado
print("Cadena de fecha y hora original:", fecha_hora_str)
print("Objeto de fecha y hora resultante:", fecha_hora_obj)
```

```python
# Volviendo a nuestro dataframe

# Vamos a aplicar el cambio a todos los elementos de la columna con una list comprehension
# Y guardamos los elementos en formato fecha en una nueva columna
got_df['Ordered_date'] = [datetime.strptime(cadena_fecha, "%B %d, %Y") for date_r in got_df['Ordered']]
```

```python
# Ahora vamos a guardar el dataset con las modificaciones
got_df.to_csv("data/Game_of_Thrones_clean.csv", index=False)
```
